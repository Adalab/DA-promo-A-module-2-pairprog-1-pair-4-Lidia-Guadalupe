**10. Utilizando la columna de 'year' cread una nueva columna llamada siglo que tiene siglo-XX o siglo-XXI. Donde el a√±o sea menor que 2000 pondremos siglo-XX y el resto de los a√±os siglo-XXI.**

```python
condiciones = [
            (df_attack2['year'] < 2000), 
            (df_attack2['year'] >= 2000)
]
opciones = ['Siglo-XX', 'Siglo-XXI']
df_attack2['siglo'] = np.select(condiciones, opciones)
df_attack2.head()

#Hemos conocido 3 formas diferentes de a√±adir columnas con condiciones, con  SELECT, WHERE y esta √∫ltima.
```

```python
#Hemos decidido cambiar su posici√≥,(con .pop)para que sea m√°s proxima 
#a la columna sobre la que se han tomado los datos de referencia para las condiciones.
nueva_posicion = df_attack2.pop('siglo') 

df_attack2.insert(3, 'siglo', nueva_posicion)
df_attack2.head()
```
